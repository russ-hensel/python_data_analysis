```python
covid_words    = [ "covid", "pandemic" , 

                    "fatalities"
                    "social-distancing",
                    "coronavirus",
                    "invisible enemy",
                    "therapies",
                 ] 


def is_text_covid( a_string ):
    """
    see if tweet is about the virus
    """
    global covid_words
    #print( a_string )
    norm_string   = a_string.lower()
    
    for a_word in covid_words:
        if norm_string.find(  a_word ) > 0:
            print( "covid" )
            return 10

    return 0
 
def test_row_apply( a_row ):
    #print( a_row )
    #print( f"type is >>>>>>>>>>{type(a_row)}<" )
    #print( f"{a_row['text']}")
    a_string         = a_row["text"]
    how_covid        = is_text_covid( a_string )
    a_row["is_covid"]  = how_covid
    #print( a_row )
    #print (f"covid>>>{is_text_covid(a_string)}<<<")
    return a_row 

print( "cell ran", datetime.datetime.now() )
```

```python
new_df = df_trump.apply( test_row_apply, axis = 1, )  
#display(df_trump.head( 20 ))
display(new_df.head( 20 ))
```
